```python
p_df=pd.read_excel('AgentPerformance (1).xlsx', skiprows=1)
p_df.head()
```

```python
p_df.info()
```

```python
# Total query that you hvae taken 
p_df.groupby('Agent Name')['Total Chats'].sum()
```

```python
active_p=p_df[p_df['Total Chats'] !=0]
```

```python
# 4. total Feedback that you have received 
active_p.groupby('Agent Name')['Total Feedback'].sum()
```

```python
# 1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
Rating_Avg=pd.DataFrame(active_p.groupby('Agent Name')['Average Rating'].mean())
Rating_Avg
```

```python
# 5. a agent name who have average rating between 3.5 to 4 
Rating_Avg[(Rating_Avg['Average Rating'] >=3.5 )& (Rating_Avg['Average Rating'] <=4)]
```

```python
# 6. a agent name who have average rating lesss then 3.5 
Rating_Avg[(Rating_Avg['Average Rating'] <3.5 )]
```

```python
# 7 . agent name who have rating more then 4.5 
Rating_Avg[(Rating_Avg['Average Rating'] >4.5 )]
```

```python
# 8 . how many feedaback agents have received more then 4.5 average
p_df[p_df['Average Rating']>4.5].groupby('Agent Name')['Total Feedback'].sum()
```

```python
# 11 . list of all agents name 
for i in p_df['Agent Name'].unique():
    print(i)
```

```python
# 9 . average weekly response time for each agent 
active_p['Date']= pd.to_datetime(active_p['Date'])
active_p['Week']=active_p['Date'].dt.week
```

```python
active_p['Average Response Sec']=(pd.to_timedelta(active_p['Average Response Time']))
active_p['Average Response Sec']=active_p['Average Response Sec'].dt.seconds
```

```python
pd.to_timedelta(active_p.groupby(['Agent Name', 'Week'])['Average Response Sec'].sum(), unit='s')
```

```python
# 10 . average weekely resolution time for each agents 
active_p['Average Resolution Sec']=(pd.to_timedelta(active_p['Average Resolution Time']))
active_p['Average Resolution Sec']=active_p['Average Resolution Sec'].dt.seconds
```

```python
pd.to_timedelta(active_p.groupby(['Agent Name', 'Week'])['Average Resolution Sec'].sum(), unit='s')
```

```python
# 12 . percentage of chat on which they have received a feedback 
feedback_percent=pd.DataFrame((active_p.groupby('Agent Name')[['Total Chats', 'Total Feedback']].sum()))
```

```python
(feedback_percent['Total Feedback']/feedback_percent['Total Chats'])*100
```
